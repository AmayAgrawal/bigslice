```python
import pandas as pd
```

```python
missings = pd.read_csv("./tmp/missing_pfams.txt", header=None)[0].str.lstrip().str.rstrip()
```

```python
all_pfams = []
import gzip
with gzip.open("./tmp/Pfam-A.hmm.gz", 'rt') as ii:
    buffer = {}
    for line in ii:
        if line.startswith("NAME"):
            buffer["name"] = " ".join(line.rstrip("\n").split()[1:])
        elif line.startswith("ACC"):
            buffer["acc"] = " ".join(line.rstrip("\n").split()[1:])
        elif line.startswith("DESC"):
            buffer["desc"] = " ".join(line.rstrip("\n").split()[1:])
        elif line.startswith("//"):
            if "desc" in buffer:
                all_pfams.append(buffer)
                buffer = {}
all_pfams = pd.DataFrame(all_pfams)
all_pfams.index = all_pfams["acc"].str.split(".").str[0]
```

```python
from_old_biosyn = pd.read_csv("./biopfam-v1.tsv", sep="\t")
from_old_biosyn.index = from_old_biosyn["Acc"].str.split(".").str[0]
```

```python
new_biosyn = pd.DataFrame({
    "Acc": all_pfams["acc"].values,
    "Name": all_pfams["name"].values,
    "Desc": all_pfams["desc"].values,
    "Status": from_old_biosyn.reindex(all_pfams.index)["Status"].fillna("").values
}, index = all_pfams.index)
new_biosyn.to_csv("./new_biosyn_pfam.tsv", sep="\t", index=False)
```

```python
new_biosyn_curated = pd.read_csv("./biosynthetic_pfams/biopfam.tsv", sep="\t")
new_biosyn_curated.index = new_biosyn_curated["Acc"].str.split(".").str[0]
new_biosyn_curated
```

```python
new_subpfam_curated = pd.read_csv("./sub_pfams/corepfam.tsv", sep="\t")
new_subpfam_curated.index = new_subpfam_curated["Acc"].str.split(".").str[0]
new_subpfam_curated["Acc"] = new_biosyn_curated.loc[new_subpfam_curated.index]["Acc"].values
new_subpfam_curated
```

```python
new_subpfam_curated.to_csv("./sub_pfams/corepfam.tsv", sep="\t", index=False)
```
